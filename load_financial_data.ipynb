```python
import datetime
import yfinance as yf

ticker = "MSFT"
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2023, 12, 31)

msft_data = yf.Ticker(ticker)
msft_df = msft_data.history(start=start_date, end=end_date)
msft_df['Close'].plot(title="MSFT's stock");
```

```python
!pip install fastparquet
```

```python
msft_df.to_parquet('msft_df.parquet', engine = 'fastparquet')
```
